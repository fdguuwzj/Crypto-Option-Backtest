```python
import pandas as pd
import numpy as np
from plotly.subplots import make_subplots
import plotly.graph_objects as go
routine_position = pd.read_csv(r'D:\program\local_repositories\option_backtester\data\output\BTC_sell_straddle_20240101_000000_20250101_000000\routine_position.csv')
routine_position['curve_delta'] = routine_position['curve'].astype(np.float64).pct_change()
routine_position['curve_delta'].fillna(0, inplace=True)
routine_position
```

```python
IV = pd.read_csv(r'D:\program\local_repositories\option_backtester\research\BTC_iv_index_with_percentile.csv')
IV
```

```python
routine_position = pd.merge(routine_position, IV, left_on='current_time', right_on='hour')
routine_position
```

```python
# routine_position['iv_percentile_ma'] = routine_position['iv_percentile'].rolling(12).mean()
routine_position['iv_signal'] = routine_position['iv_percentile'].rolling(24).max()
rotation_cond = routine_position['iv_signal'] <= 0.8

routine_position['curve_delta'] = np.where(rotation_cond, routine_position['curve_delta'], 0)
suffix = 'new'
routine_position[f'{suffix}_curve'] = (1+routine_position['curve_delta']).cumprod()
routine_position[f'{suffix}_max2here'] = routine_position[f'{suffix}_curve'].expanding().max()
# 计算到历史最高值到当日的跌幅,draw-down
routine_position[f'{suffix}_dd2here'] = routine_position[f'{suffix}_curve'] / routine_position['max2here'] - 1
routine_position
```

```python
# 创建图形
fig = make_subplots(
    rows=1, cols=1, shared_xaxes=True, vertical_spacing=0.02,
    specs=[
        [{"type": "xy", "secondary_y": True}],

    ],
)
# 添加高波动率散点 - 使用主轴
fig.add_trace(go.Scatter(
    x=routine_position['current_time'],
    y=routine_position['curve'],
    mode='lines',
    name='Curve',
    marker=dict(color='green', size=5)
))# 添加高波动率散点 - 使用主轴
fig.add_trace(go.Scatter(
    x=routine_position['current_time'],
    y=routine_position[f'{suffix}_curve'],
    mode='lines',
    name=f'{suffix}_Curve',
    marker=dict(color='blue', size=5)
))
fig.add_trace(go.Scatter(
    x=routine_position['current_time'],
    y=routine_position['iv_signal'],
    mode='lines',
    name='iv_signal',
    marker=dict(color='red', size=5)
),secondary_y=True)

fig.add_trace(go.Scatter(
    x=IV['hour'],
    y=IV['iv'],
    mode='lines',
    name='mark_iv',
    marker=dict(color='yellow', size=5)
),secondary_y=True)

fig.add_trace(go.Scatter(
    x=IV['hour'],
    y=IV['iv_percentile'],
    mode='lines',
    name='iv_percentile',
    marker=dict(color='yellow', size=5)
),secondary_y=True)
# 更新布局
fig.update_layout(
    title=f'资金曲线与最大回撤 (红色: 高波动率, 蓝色: 低波动率)',
    xaxis_title='时间',
    yaxis_title='净值',
    yaxis2=dict(
        title='回撤',
        overlaying='y',
        side='right'
    ),
    showlegend=True
)
import plotly.io as pio
pio.renderers.default = 'browser'  # 或尝试其他渲染模式
fig.show()
```
